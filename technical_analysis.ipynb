```python
# create another plot of whether people consider the drinking water safe, have
# it on the same plot as the Drinking water sources histogram, with a column for
# total people using the water source for drinkingh, and a column for the number
# of people who consider it safe.

# create a new dataframe with just the columns we want
df2 = df[['Drinking Water Supply', 'Consider drinking water safe?']]
# drop all rows where either column has a NaN value
df2 = df2.dropna(axis=0, how='any')
# get a list of the water sources people are drinking from
water_sources_drinking = df2['Drinking Water Supply'].unique().tolist()

# Create a list of zeros the same length as the water_sources list
safe_water_sources_drinking_count = [0] * len(water_sources_drinking)

# Count the number of times each water source appears in the first column and has "TRUE" in the second column
for index, row in df2.iterrows():
    if row['Consider drinking water safe?'] == True:
        safe_water_sources_drinking_count[water_sources_drinking.index(row['Drinking Water Supply'])] += 1

# Print the count for each drinkable water source
for water_source, count in zip(water_sources_drinking, drinking_count):
    print(f"{water_source}: {count}")

# Print the count for each water source
for safe_water_source, count in zip(water_sources_drinking, safe_water_sources_drinking_count):
    print(f"{safe_water_source}: {count}")

# plot this new data overlapping with the Drinking Water Sources Histogram
plt.bar(water_sources_drinking, safe_water_sources_drinking_count, color='red')
plt.title('Safe drinking water sources histogram')
plt.xlabel('Drinking Water Source')
plt.ylabel('Number of households who consider water safe')
plt.xticks(rotation=90)
plt.grid(axis='y')
# show the count of each water source above the bar, with the text centered and
# 1 pixels of space between the bar and the text
for i in range(len(water_sources_drinking)):
    plt.text(x=i, y=safe_water_sources_drinking_count[i]+1, s=safe_water_sources_drinking_count[i],
             horizontalalignment='center')

plt.show()
```

```python
bar_width = 0.35

# Create a figure and axis
fig, ax = plt.subplots()

source_bars = ax.bar(
  range(len(water_sources_drinking)),
  drinking_count,
  width=bar_width,
  label = "Drinking Water Sources",
  color = "blue"
)

drinking_bars = ax.bar(
  [i + bar_width for i in range(len(water_sources_drinking))],
  safe_water_sources_drinking_count,
  width=bar_width,
  label = "Safe Drinking Water Sources",
  color = "red"
)

# Set x-axis labels and title
ax.set_xticks([i + bar_width / 2 for i in range(len(water_sources_drinking))])
ax.set_xticklabels(water_sources_drinking)
ax.set_xlabel('Drinking Water Source')
ax.set_ylabel('Count')
ax.set_title('Available Vs Safe Drinking Water Sources')
ax.set_ylim(0,55)
ax.grid(axis='y')

# rotate axis labels
plt.setp(ax.get_xticklabels(), rotation=45)

# add counts above the two bar graphs
for i in range(len(water_sources_drinking)):
    plt.text(x=i, y=drinking_count[i]+1, s=drinking_count[i],
             horizontalalignment='center')
    plt.text(x=i+bar_width, y=safe_water_sources_drinking_count[i]+1, s=safe_water_sources_drinking_count[i],
             horizontalalignment='center')

ax.legend()

plt.show()

```
